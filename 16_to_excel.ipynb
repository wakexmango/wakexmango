```python
import pandas as pd
import numpy as np
```

```python
product_df = pd.read_csv('/content/drive/MyDrive/data/products.csv', sep=',')
product_df.head()
# product_df.tail()
```

```python
agg_prd_df = pd.pivot_table(product_df, index='product_type', columns='size', values='quantity', aggfunc=sum)
agg_prd_df
```

```python
agg_prd_df.to_excel('/content/drive/MyDrive/data/agg_prd_df.xlsx')
```

```python
agg_prd_df = agg_prd_df.reset_index()
agg_prd_df
```

```python
agg_prd_df.to_excel('/content/drive/MyDrive/data/agg_prd_df2.xlsx')
```

```python
agg_prd_df.to_excel('/content/drive/MyDrive/data/agg_prd_df3.xlsx', index=None, sheet_name='시트1입니다.')
```
