# 함수

```python
def python_study(a, b):
  print('+', a+b)
  print('*', a*b)
  print('**', a**b)
  print('/', a/b)
  print('%', a%b)
  print()
```

```python
python_study(1, 2)
python_study(2, 3)
python_study(3, 4)
```
