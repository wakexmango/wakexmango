```python
class error_msg(Exception):
  pass
```

```python
def question(answer):
  if answer.lower() != 'python':
    raise error_msg()
  print('현재 배우고 있는 언어는', answer, '입니다.')
```

```python
try:
  # question('python')
  question('java')
except error_msg:
  print('틀렸습니다.')
```
