# 딕셔너리

```python
stu = { 'name' : ['Susan', 'Jessica', 'John', 'Michael'] , 'age' : [19, 15, 17, 13] }
print(stu)
print(type(stu))
```

```python
stu = { 'name' : ['Susan', 'Jessica', 'John', 'Michael'] , 'age' : [19, 15, 17, 13, 12] }
print(stu)
print(type(stu))
```

```python
stu['math'] = [50, 100, 70, 80]
print(stu)
```

```python
del stu['math']
print(stu)
```

```python
for key, value in stu.items():
  print(key, value)
```

```python
import time

for key, value in stu.items():
  print(key, value)
  time.sleep(3)
```
