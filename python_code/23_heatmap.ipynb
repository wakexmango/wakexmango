```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head(3)
```

```python
target_df = pd.DataFrame(data=iris.target, columns=['target'])
target_df.head(3)
```

```python
target_df.loc[target_df['target'] == 0, 'target'] = 'setosa'
target_df.loc[target_df['target'] == 1, 'target'] = 'versicolor'
target_df.loc[target_df['target'] == 2, 'target'] = 'virginica'
target_df.head(3)
```

```python
df = pd.concat([iris_df, target_df], axis=1)
df.head()
```

```python
df.loc[(df['target'] == 'setosa')]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
df.set_index('target', inplace=True)
df.head(3)
```

```python
df.loc[df.index == 'setosa']
```

```python
setosa_df = df.loc[(df.index == 'setosa')]
setosa_df.head()
```

```python
plt.figure(figsize=(15, 15))
ax = sns.heatmap(setosa_df, annot=True, linewidths = 1)
```

```python
# https://matplotlib.org/3.2.1/tutorials/colors/colormaps.html

plt.figure(figsize=(15, 15))
ax = sns.heatmap(setosa_df, annot=True, fmt='f', linewidths = 1, cmap='Pastel2')
```

```python
plt.figure(figsize=(15, 15))
ax = sns.heatmap(setosa_df, annot=True, fmt='f', linewidths = 1, cmap='PuBu')
```
