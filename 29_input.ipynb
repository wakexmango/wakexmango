# for 문

```python
student = ['Susan', 'Jessica', 'John', 'Michael']
```

```python
import time

for i in student:
  print(i)
  time.sleep(2)
```

```python
len(student)
```

```python
range(len(student))
```

```python
list(range(len(student)))
```

```python
for i in range(len(student)):
  print(i)
  time.sleep(2)
```

```python
for i in range(len(student)):
  print(student[i], student[i+1])
```

```python
for i in range(len(student)-1):
  print(student[i], student[i+1])
```

```python
student
```

```python
for i in student:
  print(i)
  time.sleep(2)
```

```python
for i in student:
  print(i)
  time.sleep(2)
  
  if i == 'Jessica':
    print('Jessica님 안녕하세요.')
  elif i == 'Susan':
    print('Susan님 안녕하세요.')
  else:
    print('안녕하지 못해요.')
  
  time.sleep(2)
```
